```python
# Checking daily mean
minute_weather.index = pd.to_datetime(minute_weather.index, unit='s')
dailymean = minute_weather.resample("D").mean()
```

```python
# Looking the data computed by daily mean

dailymean.head()
```

```python
ten_daymean = minute_weather.resample("10D").mean()
ten_daymean.head()
```

```python
# Sampling every 10 days

ten_daymean = minute_weather.loc[:, "air_pressure"].resample("10D").mean()
ten_daymean.head()
```

```python
# Sum of air pressure data evry 10 days
minute_weather.loc[:, "air_pressure"].resample("10D").count().head()
```

```python
# Getting maximum temperatures for every day in september between years 2011 to 2014

september = minute_weather['avg_wind_speed'].loc['2011-09':'2014-09']
september.resample("D").max().head()
```

```python
# Getting minimum temperatures for every day in september between years 2011 to 2014

september = minute_weather['avg_wind_speed'].loc['2011-09':'2014-09']
september.resample("D").min().head()
```

```python
# Sampling last day of every year

last_daymean = minute_weather.air_pressure.resample("A").last()
last_daymean
```
